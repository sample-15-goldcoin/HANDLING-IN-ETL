```python
import pandas as pd
import numpy as np

data = {
    "Customer_ID": [101,102,103,104,105,106,107,108],
    "Name": ["Rahul Mehta","Anjali Rao","Suresh Iyer","Neha Singh",
             "Amit Verma","Karan Shah","Pooja Das","Riya Kapoor"],
    "City": ["Mumbai","Bengaluru","Chennai","Delhi",
             "Pune","Ahmedabad","Kolkata","Jaipur"],
    "Monthly_Sales": [12000,np.nan,15000,np.nan,18000,np.nan,14000,16000],
    "Income": [65000,np.nan,72000,np.nan,58000,61000,np.nan,69000],
    "Region": ["West","South","South","North",np.nan,"West","East","North"]
}

df = pd.DataFrame(data)
print(df)
```

Q8. Listwise Deletion
Remove all rows where Region is missing.
 Tasks:
 Identify affected rows
 Show the dataset after deletion
 Mention how many records were lost

```python
missing_region = df[df["Region"].isna()]
print("Rows with missing Region:")
print(missing_region)
```

```python
df_q8 = df.dropna(subset=["Region"])
print("\nDataset after deletion:")
print(df_q8)
```

```python
records_lost = len(df) - len(df_q8)
print("\nRecords lost:", records_lost)
```

Q9. Imputation Handle missing values in Monthly_Sales using: Forward Fill Tasks: Apply forward fill Show before vs after values Explain why forward fill is suitable here

```python
df_q9 = df.copy()
df_q9["Monthly_Sales"] = df_q9["Monthly_Sales"].ffill()
```

```python
comparison = pd.DataFrame({
    "Before": df["Monthly_Sales"],
    "After": df_q9["Monthly_Sales"]
})
print(comparison)
```

Why forward fill is suitable

Explanation:

Forward fill uses the last known value.

Suitable when data is sequential (e.g., time, customer order, or logical sequence).

Assumes the next value is likely similar to the previous one when missing.

Q10. Flagging Missing Data Create a flag column for missing Income. Tasks: Create Income_Missing_Flag (0 = present, 1 = missing) Show updated dataset Count how many customers have missing income

```python
df_q10 = df.copy()
df_q10["Income_Missing_Flag"] = df_q10["Income"].isna().astype(int)
```

```python
print(df_q10)
```

```python
missing_count = df_q10["Income_Missing_Flag"].sum()
print("Customers with missing income:", missing_count)
```
